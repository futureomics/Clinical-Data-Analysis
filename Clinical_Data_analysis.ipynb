```python
import matplotlib.pyplot as plt
import numpy as np

# Sample clinical data
data = {
    'Treatment_A': np.random.normal(loc=50, scale=10, size=100),
    'Treatment_B': np.random.normal(loc=45, scale=8, size=100),
    'Treatment_C': np.random.normal(loc=55, scale=12, size=100)
}

# Create a box plot
plt.figure(figsize=(10, 6))
plt.boxplot(data.values(), labels=data.keys())
plt.title('Clinical Data Box Plot')
plt.xlabel('Treatment')
plt.ylabel('Value')
plt.grid(True)
plt.show()
plt.savefig('Clinical Data.png')
```
